```python
# Color Each Dot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x,y , c=colors)
plt.show()
```

```python
#Create a color array, and specify a colormap in the scatter plot:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x,y , c=colors, cmap='viridis')
plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
plt.scatter(x,y , c=colors ,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# Size
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y , s=sizes)

plt.show()
```

```python
# Alpha
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y, s= sizes, alpha=0.5)
plt.show()
```

```python
# Combine Color Size and Alpha
x= np.random.randint(100, size=(100))
y= np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes= 10*np.random.randint(100, size=(100))
plt.scatter(x,y, c=colors, s=sizes, alpha=0.5,cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

```python
# Creating Pie Charts
y= np.array([35,25,25,15])
plt.pie(y)
plt.show()
```

```python
# Labels
y = np.array([35,25,25,15])
mylab=["Apple","Banana","Dates","Mango"]
plt.pie(y, labels = mylab)
plt.show()
```

```python
# Start Angle
plt.pie(y,labels=mylab , startangle = 90)
plt.show()
```

```python
# Explode
myexp=[0.2,0.5,0,0]
plt.pie(y, labels=mylab, explode= myexp)
plt.show()
```

```python
# Shadow
plt.pie(y,labels=mylab,explode=myexp,shadow=True)
plt.show()
```

```python
# Colors
mycol= ["black","hotpink","b","#4caf50"]
plt.pie(y,labels= mylab, colors=mycol)
plt.show()
```

```python
#Legend
plt.pie(y,labels=mylab)
plt.legend()
plt.show()
```

```python
# Legend With Header
plt.pie(y, labels=mylab)
plt.legend(title = "Four Fruits:")
plt.show()
```
